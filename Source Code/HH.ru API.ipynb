```python
def get_region_city_id():
    resp = r.get("https://api.hh.ru/areas")
    js = json.loads(resp.content.decode())
    regions_js = js[0]["areas"]

    regions = dict()
    for region in regions_js:
        cities = dict()
        for city in region["areas"]:
            cities[city.get("name")] = city.get("id")
        regions[region.get("name")] = (region.get("id"), cities)

    return regions

regions_cities_ids = get_region_city_id()
```

```python
def get_regions_id():
    resp = r.get("https://api.hh.ru/areas")
    js = json.loads(resp.content.decode())
    regions_js = js[0]["areas"]

    regions = dict()
    for region in regions_js:
        regions[region.get("name")] = (region.get("id"))

    return regions

reg_ids = get_regions_id()
```

```python
def get_cities_id():
    resp = r.get("https://api.hh.ru/areas")
    js = json.loads(resp.content.decode())
    regions_js = js[0]["areas"]

    cities = dict()
    for region in regions_js:
        for city in region["areas"]:
            cities[city.get("name")] = city.get("id")

    return cities

cities_id = get_cities_id()
```

```python
def get_page(vacance: str, area=22, page=0):
    params = {
        "text": vacance,
        "area": area,
        "page": page,
        "per_page": 100,
        "search_field": ["name", "description"]
    }
    response = r.get("https://api.hh.ru/vacancies", params=params)
    info = response.content.decode()
    return info
```

```python
def get_count(vacance: str, area=22, page=0):
    params = {
        "text": vacance,
        "area": area,
        "page": page,
        "per_page": 1,
        "search_field": ["name", "description"]
    }
    response = r.get("https://api.hh.ru/vacancies", params=params)
    count = json.loads(response.content.decode())["found"]
    return count
```

```python
def get_page_with_count(vacance: str, area=22, page=0):
    params = {
        "text": vacance,
        "area": area,
        "page": page,
        "per_page": 100,
        "search_field": ["name", "description"]
    }
    response = r.get("https://api.hh.ru/vacancies", params=params)
    count = json.loads(response.content.decode())["found"]
    info = response.content.decode()
    return info
```

```python
def get_vacancie(vacance, area):
    js_objs = []
    total_pages = json.loads(get_page(vacance, area))["pages"]
    time.sleep(TIME_OUT)
    for i in range(total_pages): # Максимально в HH.ru можно получить только 2000 результатов (20 страниц по 100 элементов)
        js = json.loads(get_page(vacance, area, i))
        js_objs.extend(js["items"])
        time.sleep(TIME_OUT)
    return js_objs
```
